```python
dff=df.drop(["BldgType","CentralAir"],axis=1)
dff
```

```python
X=dff.drop(["SalePrice"],axis=1)
X
```

```python
X
```

```python
y=dff["SalePrice"].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
import sklearn.metrics as metrics
mse=metrics.mean_squared_error(y_test,y_pred)
print("MSE without Categorical: ",mse)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.title("Plot b/w y_test vs y_pred")
plt.show()
```

```python
encode=pd.get_dummies(df)
encode
```

```python
X=encode.drop(["SalePrice"],axis=1)
X
```

```python
y=encode["SalePrice"].values
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
y
```

```python
import sklearn.metrics as metrics
mse_cd=metrics.mean_squared_error(y_test,y_pred)
print("MSE with Categorical data: ",mse_cd)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
std_X_train=std.fit_transform(X_train)
std_X_train
```

```python
std_X_test=std.transform(X_test)
std_X_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(std_X_train,y_train)
std_y_pred=lr.predict(std_X_test)
std_y_pred
```

```python
std_mse=metrics.mean_squared_error(y_test,std_y_pred)
print("Mean Squared Error: ",std_mse)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
mm_X_train=mm.fit_transform(X_train)
mm_X_test=mm.transform(X_test)

reg.fit(mm_X_train,y_train)
mm_y_pred=reg.predict(mm_X_test)
mm_y_pred
```

```python
mm_mse=metrics.mean_squared_error(y_test,mm_y_pred)
print("Mean Squared Error: ",mm_mse)
```

```python
from sklearn.linear_model import SGDRegressor
sgd=SGDRegressor()
sgd.fit(std_X_train, y_train)
sgd_y_pred=sgd.predict(std_X_test)
sgd_y_pred
```

```python
sgd_mse=metrics.mean_squared_error(y_test, sgd_y_pred)
print("Mean Squared Error:",sgd_mse)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(std_X_train, y_train)
ridge_y_pred=ridge.predict(std_X_test)
ridge_y_pred
```

```python
ridge_mse=metrics.mean_squared_error(y_test, ridge_y_pred)
print("RIDGE Mean Squared Error:",ridge_mse)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(std_X_train, y_train)
lasso_y_pred=lasso.predict(std_X_test)
lasso_y_pred
```

```python
lasso_mse=metrics.mean_squared_error(y_test, lasso_y_pred)
print("LASSO Mean Squared Error:",lasso_mse)
```
